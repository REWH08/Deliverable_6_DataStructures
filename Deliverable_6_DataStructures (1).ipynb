```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

summary_dictionary = {'mean':mean, 'std':std, 'min':min, 'max':max}

summary_df = pd.DataFrame(data = summary_dictionary)

summary_df.index = (['sepal_length', 'sepal_width',	'petal_length',	'petal_width',	'sepal_sum',	'petal_sum'])

pd.set_option('display.colheader_justify', 'center')

summary_df
```
